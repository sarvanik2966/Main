```python
# Bar chart
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("D:/MAINFLOW/householdtask3.csv")

```

```python
plt.bar (data['year'], data ['income'])
plt.title("Bar Chart")
# Setting the X and Y labels
plt.xlabel ('Year' )
plt.ylabel( 'Income')
# Adding the legends
plt. show()
```

```python
# Create histograms for household income and expenditure
plt.figure(figsize=(12, 6))

# Histogram for Income
plt.subplot(1, 2, 1)
plt.hist(data['income'], bins=15, color='blue', alpha=0.7, edgecolor='black')
plt.xlabel('Income (USD)')
plt.ylabel('Frequency')
plt.title('Distribution of Household Income')

# Histogram for Expenditure
plt.subplot(1, 2, 2)
plt.hist(data['expenditure'], bins=15, color='green', alpha=0.7, edgecolor='black')
plt.xlabel('Expenditure (USD)')
plt.ylabel('Frequency')
plt.title('Distribution of Household Expenditure')

plt.tight_layout()
plt.show()
```
